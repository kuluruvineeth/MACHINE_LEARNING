```python
import math
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as rnd
import seaborn as sns
sns.set()
from scipy.stats import norm
s=np.linspace(0,10,10)
N=10
def likelihood(x):
    i=0
    sum=0
    while(i<N):
        sum+=pow((s[i]-x),2)
        i=i+1
    y=-(0.5*sum + 5*(math.log(2*3.14)))
    p=math.exp(y)
    return p
l=np.vectorize(likelihood)
x = np.linspace(0,10,10)
plt.title("likelihood curve")
plt.xlabel("mean")
plt.ylabel("likelihood values(product of probabilities)")
plt.plot(x,l(x),'b')
plt.plot(x,l(x),'ro')
plt.show()
```
