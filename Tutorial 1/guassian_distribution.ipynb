```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import numpy.random as rnd
```

```python
data=rnd.standard_normal(10)
print(np.mean(data))
print(np.std(data))
```

$\mu=5$, $sigma=1$

```python
sig=1
mu=5
data=sig * rnd.standard_normal(10) + mu
print(np.mean(data))
print(np.std(data))
```

```python
plt.hist(data)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import numpy.random as rnd
import seaborn as sns
sns.set()
from scipy.stats import norm
x = np.linspace(0,10,10)
y = norm.pdf(x,loc=5,scale=1)
plt.title("Gaussian distribution")
plt.xlabel("sample points")
plt.ylabel("probability")
plt.plot(x,y,'b')
plt.plot(x,y,'ro')
```
