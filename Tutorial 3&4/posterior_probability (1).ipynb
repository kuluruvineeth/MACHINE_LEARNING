```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
def dataset():
    x=np.linspace(-1,1,10)
    y=0.5*x-0.3
    t=y+np.random.normal(0,0.2,10)
    return t

sample_points = dataset()

likelihood_grid = norm(0,0.2).pdf(sample_points)
prior_grid = norm(0,0.71).pdf(sample_points)
likelihood_grid,prior_grid = np.meshgrid(likelihood_grid,prior_grid)

def prod(a,b):
    return np.multiply(a,b)

posterior = prod(likelihood_grid,prior_grid)
plt.figure(figsize=(12,6))
heatmap = sns.heatmap(posterior,xticklabels=True,yticklabels=True,center=0,annot=True,linewidth=0.5,cmap='BuPu')
heatmap.invert_yaxis()
```
